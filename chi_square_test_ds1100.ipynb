```python
import scipy
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
df = sns.load_dataset('tips')
H0 = "There is no relationship between sex and smoking"
H1 = "Thre is siginifcant relationship between sex and smoking."
alpha = 0.05
table = pd.crosstab(df['sex'], df['smoker'])
result = chi2_contingency(table)

O = table.values
E = result.expected_freq

chi_sq = np.sum((O - E)**2/E)
confidence_interval = 1 - alpha
degree_of_freedom = result.dof
chi_square_tabular = scipy.stats.chi2.ppf(confidence_interval, df=degree_of_freedom) # computes tabular value

if chi_sq > chi_square_tabular:
   print("We reject H0")
else:
   print("We fail to reject H0")
```
