# IQ => Primary key vs Foreign key

# Inner join -Default (Intersection data only)

```python
pd.merge(employee, department, on="DepartmentID", how="inner")
```

# Outer join (Union)

```python
pd.merge(employee, department, on="DepartmentID", how="outer")
```

# left join

```python
pd.merge(employee, department, on="DepartmentID", how="left")
```

```python
# NaN means missing value; DepartmentID 106 has no DepartmentName and Location so value is missing
```

```python
pd.merge(department, employee, on="DepartmentID", how="left")
```

# right join

```python
pd.merge(employee, department, on="DepartmentID", how="right")
```

```python
pd.merge(department, employee, on="DepartmentID", how="right")
```
