```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Types of data analysis
* Univariate: One variable
* Bivariate: Two variable
* Multi-variate : Three or more variables

# Univariate
-- line chart

```python
month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul"]
growth = [3, 3.6, 2, 2.3, -2, 1, 5]
```

```python
plt.plot(month, growth)
```

```python
plt.plot(month, growth, "o-")
```

```python
plt.plot(month, growth, "o-", color="Purple")
```

```python
plt.plot(month, growth, "o-", color="purple")
plt.grid()
plt.xlabel("Month")
plt.ylabel("Growth Percentage (%)")
```

```python
plt.plot(month, growth, "o-", color="purple")
plt.grid()
plt.xlabel("Month")
plt.ylabel("Growth Percentage (%)")
plt.show()
```

```python
month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul"]
actual_growth = [2, 3, 2, 6, 4, 5, 7]
predicted_growth = [3, 4, 4, 5, 6, 7, 8]
```

```python
plt.plot(month, actual_growth, "o-", label="actual_growth")
plt.plot(month, predicted_growth, "o--", label="predicted_growth")
plt.xlabel("Months")
plt.ylabel("Growth Percentage(%)")
plt.legend()
plt.show()
```

```python
over = [i for i in range(1, 11)]
nepal_run = [10, 15, 19, 23, 23, 27, 50, 55, 72, 90]
uae_run = [5, 12, 12, 19, 25, 25, 29, 50, 55, 70]
plt.plot(over, nepal_run, label="NEPAL")
plt.plot(over, uae_run, label="UAE")
plt.xlabel("over")
plt.ylabel("runs")
plt.legend()
plt.show()
```

# Barplot

```python
gender = ["Female", "Male"]
count = [4, 8]
```

```python
plt.bar(x=gender, height=count)
plt.xlabel("Gender")
plt.ylabel("Freqencey/Count")
plt.show()
```

# using seaborn

```python
sns.barplot(x=gender, y=count, hue=["Red", "Blue"])
plt.show()
```
