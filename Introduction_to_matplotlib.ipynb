# Stacked Barplot
- cannot be done usign piechart

```python
month = ["Jan", "Feb", "Mar", "Apr"]
male_present = [120, 113, 90, 150]
female_present = [110, 90, 90, 140]
```

```python
#stacked barplot
plt.bar(month, male_present, label="Male")
plt.bar(month, female_present, label="Female")
plt.legend()
plt.show()
```

```python
#creating DataFrame
import pandas as pd
df = pd.DataFrame({
    "Month": month,
    "male_present": male_present,
    "female_present": female_present
})

# plotting graph
df.plot(x="Month", y=["male_present", "female_present"], kind="bar")
```

# The relationship between two variables is shown by correlation and we can visualize it using Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
```

# The correlation between petal_length and petal_width is 0.96. It is positive correlation. It means when the value of petal_length increases the value of petal_width also increases. You can visualize this using scatterplot.

```python
sns.scatterplot(x="petal_length", y="petal_width", data=df, hue=df["species"])
```

```python
import numpy as np
x = np.linspace(-5, 5, 100)
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
# for [0, 0]
ax = axes[0, 0]
ax.plot(x, sinx, label="sinx")
ax.legend()
ax.set_xlabel("X")

# for [0, 1]
ax = axes[0, 1]
ax.plot(x, cosx, label="cosx")
ax.legend()
ax.set_xlabel("X")

# for [1, 0]
ax = axes[1, 0]
ax.plot(x, tanx, label="tanx" )
ax.legend()
ax.set_xlabel("X")

fig.show()
```
