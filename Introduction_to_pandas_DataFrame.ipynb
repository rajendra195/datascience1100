# Creating a DataFrame using list of list

```python
data = [["e01", "Kshtiz"], ["e02", "Shyam"], ["e03", "Rita"]]
df = pd.DataFrame(data)
df
```

# How to add column name?

```python
data = [["e01", "Kshtiz"], ["e02", "Shyam"], ["e03", "Rita"]]
df = pd.DataFrame(data)
df.columns = ["emp_id", "emp_name"]
df
```

```python
data = [["e01", "Kshtiz"], ["e02", "Shyam"], ["e03", "Rita"]]
df = pd.DataFrame(data, columns=["emp_idd", "emp_name"])
df
```
