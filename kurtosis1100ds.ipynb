# Kurtosis refers to the degree of presence of outliers in the distribution. Kurtosis is a statistical measure, whether the data is heavily-tailed or light-tailed in a normal distribution.

There are 3 types of Kurtosis:
- 1: Mesokurtic
- 2: Platykurtic
- 3: Leptokurtic

# Mesokurtic
**Normal Distribution:**
- mean = $\mu$
- variance = $\sigma^2$

**Medium Outlier**

**Kurtosis = 3**


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis
np.random.seed(0)
data = np.random.normal(loc=5, scale=1, size=1500) # mean = loc, std = scale
kurtosis = kurtosis(data, fisher=False) # Fisher = False means pearson method

plt.title(f"Kurtosis =  {round(kurtosis, 2)}")
sns.histplot(data, kde=True)
plt.show()
```

# Platykurtic
* Uniform Distribution
* No Outliers
* Kurtosis = less than 3

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
np.random.seed(0)
data = np.random.uniform(low=3, high=6, size=15000) # yo matra naya
kurtosis = kurtosis(data, fisher=False)
plt.title(f"Kurtosis = {round(kurtosis, 2)}")
sns.histplot(data, bins=3)
plt.show()
```

# Leptokurtic
* Laplace distribution
* High outliers
* kurtosis > 3

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis
np.random.seed(0)
data = np.random.laplace(loc=5, scale=1, size=1500) # yo matra naya
kurtosis = kurtosis(data, fisher=False)
plt.title(f"Kurtosis = {round(kurtosis, 2)}")
sns.histplot(data, kde=True)
plt.show()
```
